```python
from ultralytics import YOLO

#Load model
model = YOLO('yolov8m.pt')

# Train the model
model.train(data="/content/drive/MyDrive/TACO/TACO.yaml",
            epochs=100,
            patience=50,
            imgsz=1280,
            device=0,
            pretrained=True,
            optimizer='Adam',
            lr0=0.001,
            batch=8
            )

metrics = model.val()
```
